O uso da função *inner_join* no pacote **Tidyverse** é muito útil para combinar conjunto de dados. Veja, nos exemplos abaixo, como fica a combinação destes *datasets* que foram criados anteriormente.

Pense nos seguintes desafios:

**1)** Combinar o resultado de **pontuacaoJogadas** que tem a informação de quando um time fez *touchdown* (significa que marcou 6 pontos no jogo) e **descricaoJogadas** onde há uma descrição da jogada. Estes conjuntos de dados estão segmentados, cada um deles possui uma parte da informação. Ao combinar estes dois conjuntos de dados é possível ter todas as variáveis juntas como se fossem um único *dataset*. Eles se combinam a partir da variável *play_id*, que é comum entre eles. A partir desta combinação, a manipulação é similar ao que já foi estudado anteriormente.

```r
pontuacaoJogadas %>% #primeiro dataset 
inner_join(descricaoJogadas, by='play_id') %>% #segundo dataset combinando com o primeiro
select(play_id, posteam, touchdown, td_team, desc) %>% #seleção de variáveis
filter(touchdown == 1) #filtro de dados
```

**2)** Mostrar qual foi o jogador do time da casa e quando ele recebeu o primeiro passe que permitiu correr 5 ou mais jardas.

Para isso, é necessário cobinar 3 conjuntos de dados. No dataset **jogo** é possível retornar qual é o time da casa. Já em **acoesJogadas** é possível saber quantas jardar foram conquistadas (com a variável *yards_gained*). E por fim, em **pontuacaoJogadas** há o nome de quem correu com a bola (variável *rusher*). Vamos ver como fica essa combinação?

```r
jogo %>% #primeiro dataset 
inner_join(acoesJogadas, by='play_id') %>% #segundo dataset combinando com o primeiro
inner_join(pontuacaoJogadas, by='play_id') %>% #terceiro dataset combinando com o primeiro e o segundo
select(play_id, home_team,posteam, rusher, yards_gained, time, qtr ) %>% #seleção de variáveis
filter( posteam == home_team | yards_gained >=5 ) %>% #filtro de dados
head(1) #retorno apenas de 1 linha
```

Utilizando o subconjunto de dados **acoesJogadas** e **pontuacaoJogadas**, crie uma análise que retorne qual foi o jogador que conquistou mais jardas no terceiro quarto.

```r
## Resultado do jogador com mais jardas no terceiro quarto, em um unico jogo ###
acoesJogadas %>% #primeiro dataset 
inner_join(pontuacaoJogadas, by='play_id') %>% #segundo dataset combinando com o primeiro
select(play_id, posteam, rush_attempt, rusher, yards_gained, qtr) %>% #seleção de variáveis
filter( rush_attempt ==1, qtr == 3, rusher != 'NA') %>% #filtro de dados
arrange(desc(yards_gained)) %>%  #orderna decrescente por yards_gained
head(2) #retorno top 2 linhas
```

```r
## Resultado do jogador com mais jardas no terceiro quarto, considerando todos os jogos ###
topJardas <- acoesJogadas %>% #primeiro dataset 
inner_join(pontuacaoJogadas, by='play_id') %>% #segundo dataset combinando com o primeiro
select(posteam,rush_attempt, rusher, qtr, yards_gained) %>% #seleção de variáveis
filter( rush_attempt ==1, qtr == 3 ) %>% #filtro de dados
group_by(rusher) %>% #group_by
mutate(somayards = sum(yards_gained)) %>% #sum
select(posteam,rusher, somayards) %>% #seleção de variáveis
unique() %>%  ## Unicos
arrange(desc(somayards)) #orderna decrescente por totalyards
head(topJardas,1) #retorno 1 linha
```

#### Desafio de geração de gráfico  

Crie um gráfico de linhas, mostrando a pontuação de cada time em cada *quarter*. O resultado deve ter duas linhas, uma para cada time, e cada linha será composta pela pontuação de cada um dos *quarters* sendo uma cor para cada time. O eixo X terá os *quarters* e o eixo y terá a pontuação.

```r
#Dataset com todos os times, temporada 2014
TodosTimes <- temporada 

tabDados <- unique(TodosTimes %>%
    select(home_team, home_score,away_team, away_score, total,qtr,yards_gained))
```

```r
### RESULTADOS JOGANDO EM CASA
resultCasa <- tabDados  %>%
              filter( home_team == 'SEA'
                    | home_team =='ARI'
                    | home_team == 'CAR'
                    | home_team == 'DAL'
                    | home_team == 'DEN'
                    | home_team == 'GB'
                    | home_team == 'LA' 
                    | home_team == 'LV' 
                    | home_team == 'NE' 
                    | home_team == 'NYG'
                    | home_team == 'SEA'
                    | home_team == 'SF'
                    | yards_gained != 'NA'
                    | qtr != 5
              ) %>%
              select("home_team", "home_score", "qtr","yards_gained")

### RESULTADOS JOGANDO EM FORA
resultFora <- tabDados  %>%
             filter( away_team == 'SEA'
                    | away_team =='ARI'
                    | away_team == 'CAR'
                    | away_team == 'DAL'
                    | away_team == 'DEN'
                    | away_team == 'GB'
                    | away_team == 'LA' 
                    | away_team == 'LV' 
                    | away_team == 'NE' 
                    | away_team == 'NYG'
                    | away_team == 'SEA'
                    | away_team == 'SF'
                    | yards_gained != 'NA'
                    | qtr != 5
              ) %>%
              select("away_team", "away_score", "qtr","yards_gained")

### NORMALIZANDO DATASET CASA
resultCasa <- resultCasa  %>%
              rename (Time= home_team,
                      Pontos= home_score,
                      Quarter= qtr,
                      Jardas= yards_gained)

### NORMALIZANDO DATASET FORA
resultFora <- resultFora  %>%
              rename (Time= away_team,
                      Pontos= away_score,
                      Quarter= qtr,
                      Jardas= yards_gained)

### JUNTANDO OS RESULTADOS
DadosFinaisGraf <- bind_rows(resultCasa, resultFora)
```

```r
## Dataset AGRUPANDO TIME E QUARTER, PARA SOMAR OS PONTOS
dadosagrupados <- unique(DadosFinaisGraf  %>%
    select(Time, Pontos, Quarter)  %>% #seleção de variáveis
    filter(Quarter !=5) %>% #filtro de dados
    group_by(Time, Quarter) %>% #group_by
    mutate(somaPontos = sum(Pontos))) #sum
```

```r
library(ggplot2)

## DATASET TIME SEA E GB
dadosSEAGB <- dadosagrupados  %>%
              filter( Time == 'SEA' | Time == 'GB')
ggplot(dadosSEAGB) + aes(Quarter, somaPontos) + geom_line(aes(colour = Time, linetype = Time)) 

## DATASET TODOS OS TIMES
ggplot(dadosagrupados) + aes(Quarter, somaPontos) + geom_line(aes(colour = Time, linetype = Time)) 
```

Crie um gráfico de barras empilhada (colunas verticais), utilizando somente as jogadas que tiveram entre 10 e 20 jardas conquistadas.
O empilhamento das barras será feito pela quantidade de jardas conquistadas (entre 10 e 20). Mantenha as barras verticais segmentadas por quarter do jogo, e por fim, crie a faceta baseada nos times.


```r
## GRÁFICOS
library(ggplot2)

## Dataset AGRUPANDO TIME E QUARTER, PARA SOMAR AS JARDAS
dadosJardasSomadas <- unique(DadosFinaisGraf  %>%
               filter( Jardas >=10, Jardas <=20, Quarter !=5 )  %>%
               select(Time, Quarter, Jardas) %>% #seleção de variáveis
               group_by(Time, Quarter) %>% #filtro de dados
               mutate(somaJardas = sum(Jardas))) #sum
              

## JARDAS CONQUISTADAS SOMADAS,  SEGMENTADO POR QUARTER
ggplot(dadosJardasSomadas) +  geom_col(aes(x = Time, y = somaJardas, fill = Quarter))


## GRAFICO JARDAS x TIMES
##ggplot(dadosJardas) +  geom_col(aes(x = Time, y = Jardas, fill = Time))
```
